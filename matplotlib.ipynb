```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use("dark_background")
```

```python
x = np.arange(10)
y = x**2
y1 = 2*x + 4
```

```python
# line plot
plt.plot(x,y,label = "apple",color = "red",linestyle = "dashed",marker = "o")
plt.plot(x,y1,label = "banana",color = "blue",marker = "*")
plt.xlabel("time")
plt.ylabel("price")
plt.title("line plot")
plt.legend()
plt.show()
```

```python
# ploting one axis data only
# passed asis is auto taken as y axis
# x is auto taken as index values
plt.plot(y)
plt.show()
```

```python
# scatter plot
plt.figure(figsize = (8,5))
plt.scatter(x,y,label = "scatter",marker = "o")
plt.xlabel("time")
plt.ylabel("prices")
plt.legend()
plt.title("scatter plot")
plt.show()
```

one dimentional data

```python
# bar graph
#x = np.array([1,2,3]) # distance where each bar is going to located
# to make them far apart
x = np.array([2,4,6])
y = [12,17,10]
y1 = [5,20,16]
ticks = ["gold","platinum","silver"]
plt.bar(x,y1,tick_label = ticks,width = .5,label = "year1",color = "orange")
# these graphs will overlap here
# plt.bar(x,y,tick_label = ticks)
plt.bar(x+.5,y,width = .5,label = "year2",color = "red")
# we can make it more spaced form four sides by specifying the limts
plt.xlim(1,8)
plt.ylim(0,30)
plt.legend()
plt.show()
```

```python
# pie chart
plt.figure(figsize = (5,9))
subjects = ["maths","chem","physics","english"]
weitages = [20,10,15,5]
# explode used to remove one piece or more
# autopct is to show percentage and format of percentage
plt.pie(weitages,labels = subjects,explode = (0,0,.2,0),autopct = "%1.1f%%" )
plt.title("marks and thier weitage")
# plt.legend()
plt.show()
```

distributions using histograms

```python
xsn = np.random.randn(100) # gives 100 values on standard normal distribution
sigma = 5
u = 15
X = xsn*sigma + u
X = np.round(X) # to make data intigral
X2 = xsn*15 + 50
# alpha specifies opacity values
plt.hist(X,alpha = .3)
plt.hist(X2,alpha = .4)
plt.show()
```
